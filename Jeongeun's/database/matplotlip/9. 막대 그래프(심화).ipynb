## 9. 막대 그래프(심화)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  #Windows
matplotlib.rcParams['font.size'] = 15  #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
labels = ['강백호','서태웅','정대만'] #이름
values = [190,187,184] #키

plt.barh(labels,values)
plt.xlim(175,195)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('//') # ////
bar[1].set_hatch('x') # xxxx
bar[2].set_hatch('..') # ....
```

```python
bar= plt.bar(labels, values)
plt.ylim(175,195)

for idx, rect in enumerate(bar):   #idx:0,1,2   rect: 각 bar에 대한 크기나 위치 정보
    plt.text(idx,rect.get_height()+0.5, values[idx],ha='center',color='blue')  #rect.get_height()를 통해 높이를 가져옴
```
